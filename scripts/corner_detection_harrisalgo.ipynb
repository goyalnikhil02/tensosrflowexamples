```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

flat_chess=cv2.imread('D://WebCamMotionDetector//images//chess.png')
plt.imshow(flat_chess)
```

```python
flat_chess_gray=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)

plt.imshow(flat_chess_gray,cmap='gray')

#dst=cv2.cornerHarris(src=flat_chess_gray,blockSize=2,ksize=3,k=0.04)

dst = cv2.cornerHarris(flat_chess_gray,2,3,0.04)


dst=cv2.dilate(dst,None)

flat_chess[dst>0.01*dst.max()]=[0,255,0]

plt.imshow(flat_chess)

```
