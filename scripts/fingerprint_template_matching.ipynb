```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


#data_full=cv2.imread('D://WebCamMotionDetector//images/test.jpg')

data_full=cv2.imread('D://WebCamMotionDetector//images//FingerImages//full.jpg')



# Convert it to grayscale 
data_full=cv2.cvtColor(data_full,cv2.COLOR_BGR2RGB)



#print(data_full.shape)
plt.imshow(data_full)


#face_data=cv2.imread('D://WebCamMotionDetector//images//face.png')



face_data=cv2.imread('D://WebCamMotionDetector//images//FingerImages//finger0.png')

face_data=cv2.cvtColor(face_data,cv2.COLOR_BGR2RGB)
#plt.imshow(face_data)
#print(face_data.shape)


#https://www.geeksforgeeks.org/template-matching-using-opencv-in-python/


methods=['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR','cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED']

for m in methods:
    full_copy=data_full.copy()
    method=eval(m)
    res=cv2.matchTemplate(full_copy,face_data,method)
    min_val,max_val,min_loc,max_loc=cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        top_left=min_loc
    else:
        top_left=max_loc
    height,width,channel=face_data.shape
    bottom_right=(top_left[0]+width,top_left[1]+height)
    
    cv2.rectangle(full_copy,top_left,bottom_right,(255,0,0),10)
    plt.subplot(121)
    plt.imshow(res)
    plt.title('Heat map of template matching')
    
    
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title('Detection of the template')
    #title useds
    plt.suptitle(m)
    plt.show()
    print('\n')
    print('\n')
        


#my_method=eval(methods[0])

#res=cv2.matchTemplate(data_full,face_data,my_method)
#plt.imshow(res)    
    
```
